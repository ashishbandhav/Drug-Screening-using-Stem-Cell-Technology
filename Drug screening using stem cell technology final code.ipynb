## Model optimization

### Subtask:
Optimize the hyperparameters of the trained regression models using the validation set.


**Reasoning**:
Optimize the hyperparameters of the trained regression models using GridSearchCV.



```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, mean_squared_error

# Define parameter grids for each model
param_grids = {}

param_grids['RandomForest'] = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

param_grids['SVR'] = {
    'C': [0.1, 1, 10],
    'epsilon': [0.01, 0.1, 1]
}

param_grids['GradientBoosting'] = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1],
    'max_depth': [3, 5, 7]
}

# Define scoring metric
scoring = make_scorer(mean_squared_error, greater_is_better=False)

# Perform hyperparameter optimization for each model
best_estimators = {}
for name, model in models.items():
    print(f"Optimizing {name}...")
    grid_search = GridSearchCV(model, param_grids[name], scoring=scoring, cv=5)
    try:
        grid_search.fit(X_train_scaled.drop(columns=['S_id', 'filedir', 'label', 'feature_SIFT', 'Area']), y_train)
        best_estimators[name] = grid_search.best_estimator_
        print(f"Best hyperparameters for {name}: {grid_search.best_params_}")
        print(f"Best score for {name}: {grid_search.best_score_}")
    except ValueError as e:
        print(f"Error during optimization of {name}: {e}")
```

## Model evaluation

### Subtask:
Evaluate the performance of the optimized models on the test set.


**Reasoning**:
Evaluate the performance of the optimized models on the test set using MSE, R-squared, and MAE.



```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Columns to drop for prediction
cols_to_drop = ['S_id', 'filedir', 'label', 'feature_SIFT', 'Area']

evaluation_metrics = {}
for name, model in best_estimators.items():
    print(f"Evaluating {name}...")
    try:
        y_pred = model.predict(X_test_scaled.drop(columns=cols_to_drop))

        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)

        evaluation_metrics[name] = {'MSE': mse, 'R-squared': r2, 'MAE': mae}
        print(f"{name}: MSE={mse:.2f}, R-squared={r2:.2f}, MAE={mae:.2f}")

    except ValueError as e:
        print(f"Error during evaluation of {name}: {e}")

# Print summary table
print("\nSummary of Evaluation Metrics:")
for name, metrics in evaluation_metrics.items():
    print(f"{name}:")
    for metric, value in metrics.items():
      print(f"  {metric}: {value:.2f}")
```

## Data visualization

### Subtask:
Visualize the model performance and key findings.


**Reasoning**:
Visualize the model performance using histograms, box plots, scatter plots, and a summary table, and compare model performance using bar charts.  Also, create a feature importance plot for tree-based models.



```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np  # Required for feature importance sorting

# Columns to drop for prediction
cols_to_drop = ['S_id', 'filedir', 'label', 'feature_SIFT', 'Area']

# Drop the columns once for reuse
X_input = X_test_scaled.drop(columns=cols_to_drop)

# Create a figure for the plots
plt.figure(figsize=(20, 15))

# 1. Histogram: Actual vs Predicted Values
plt.subplot(2, 2, 1)
plt.hist(y_test, bins=30, alpha=0.5, label='Actual')

for name, model in best_estimators.items():
    y_pred = model.predict(X_input)
    if len(y_pred) == len(y_test):
        plt.hist(y_pred, bins=30, alpha=0.5, label=name)
    else:
        print(f"Skipping histogram for {name} due to length mismatch.")

plt.xlabel('Differentiation Efficiency Index')
plt.ylabel('Frequency')
plt.title('Distribution of Actual vs Predicted Values')
plt.legend()

# 2. Box Plot: Actual vs Predicted Values
# Prepare prediction DataFrame
predictions = {}
for name, model in best_estimators.items():
    y_pred = model.predict(X_input)
    if len(y_pred) == len(y_test):
        predictions[name] = y_pred
    else:
        print(f"Skipping boxplot for {name} due to length mismatch.")

plot_df = pd.DataFrame({'Actual': y_test})
for name, y_pred in predictions.items():
    plot_df[name] = y_pred

plt.subplot(2, 2, 2)
sns.boxplot(data=plot_df)
plt.title('Box Plots of Actual vs Predicted Values')

# 3. Scatter Plot: Actual vs Predicted
plt.subplot(2, 2, 3)
for name, model in best_estimators.items():
    y_pred = model.predict(X_input)
    if len(y_pred) == len(y_test):
        plt.scatter(y_test, y_pred, label=name, alpha=0.5)
    else:
        print(f"Skipping scatter plot for {name} due to length mismatch.")

plt.xlabel('Actual Differentiation Efficiency Index')
plt.ylabel('Predicted Differentiation Efficiency Index')
plt.title('Scatter Plots of Actual vs Predicted Values')
plt.legend()

# 4. Bar Plot: Evaluation Metrics
summary_df = pd.DataFrame(evaluation_metrics).T
display(summary_df)
summary_df.to_csv('model_evaluation_metrics.csv')

plt.subplot(2, 2, 4)
summary_df.plot(kind='bar', ax=plt.gca())
plt.title("Model Comparison")
plt.ylabel("Metric Value")
plt.xticks(rotation=45)
plt.tight_layout()

# 5. Feature Importance for Tree-based Models
for name, model in best_estimators.items():
    if name in ['RandomForest', 'GradientBoosting']:
        plt.figure(figsize=(10, 6))
        importances = model.feature_importances_
        indices = np.argsort(importances)[::-1]
        plt.title(f"Feature Importance for {name}")
        plt.bar(range(X_input.shape[1]), importances[indices], align='center')
        plt.xticks(range(X_input.shape[1]), X_input.columns[indices], rotation=90)
        plt.tight_layout()
        plt.savefig(f'{name}_feature_importance.png')

plt.show()
```
